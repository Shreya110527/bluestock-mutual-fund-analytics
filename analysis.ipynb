```python
import pandas as pd
import os

data_path = "data/raw"
files = sorted(os.listdir(data_path))

summary = []

for file in files:
    df = pd.read_csv(os.path.join(data_path, file))

    summary.append({
        "Dataset": file,
        "Rows": df.shape[0],
        "Columns": df.shape[1]
    })

summary_df = pd.DataFrame(summary)

print(summary_df)
```
